### Pandas应用案例：电影数据分析

#### （1）从文件读出数据加载到数据帧

```python
import pandas as pd

df = pd.read_excel('电影信息.xlsx')
df
```

#### （2）提取全部的演员和电影一一对应关系

```python
pairs = []
for i in range(len(df)):
    actors = df.at[i, '演员'].split('，')
    for actor in actors:
        pair = (actor, df.at[i, '电影名称'])
        pairs.append(pair)
        
pairs = sorted(pairs, key=lambda item: int(item[0][2:]))
pairs
```

#### （3）统计每位演员参演电影的数量，修改列名，并按参演数量排序

```python
index = [item[0] for item in pairs]
data = [item[1] for item in pairs]
dfl = pd.DataFrame({'演员':index, '电影名称':data})
result = dfl.groupby('演员', as_index=False).count()
result.columns = ['演员', '参演电影数量']
result = result.sort_values(by='参演电影数量', ascending=False)
result
```

#### （4）找到参演电影数量最多的3位演员

```python
result.nlargest(3, '参演电影数量')
```

#### （5）数据可视化

```python
import matplotlib.pyplot as plt
from pylab import mpl

%matplotlib inline

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

plt.figure()
result.plot(x='演员', kind='bar')
plt.show()
```
